```python
# QUESTÃO 2A
# ALUNO: JOAO GABRIEL DA SILVA, MATRICULA 202504009

#instalar numpy via pip install
import numpy as  np #importando a biblioteca

while True: #o progama começa com um loop while True que solicita que o usuario o valore do orçamento para a categoria Essencial.
    try:
        orçamento_es = float(input("Digite o valor (em reais) que sera destinado para as despesas essenciais: R$")) #o valor e salvo como um float
        orçamento_es_a = round(orçamento_es, 2)  #arredonda o valor fornecido
        print("")
        break #encerra o loop
    except ValueError: # se a conversão para float falhar e exibida uma mensagem de erro e o loop e reiniciado.
        print("Valor invalido. Por favor tente novamente.")
        print("")

while True: #loop while True que solicita que o usuario o valore do orçamento para a categoria Lazer
    try:
        orçamento_la = float(input("Digite o valor (em reais) que sera destinado para as despesas de lazer: R$")) # a variavel armazena o orçamento em float.
        orçamento_la_a = round(orçamento_la, 2) #o valor e arredondado 
        print("") 
        break #encerra o loop
    except ValueError: # se a conversão para float falhar e exibida uma mensagem de erro e o loop e reiniciado.
        print("Valor invalido. Por favor tente novamente.")
        print("")
```

```python
# QUESTÃO 2B
# ALUNO: JOAO GABRIEL DA SILVA, MATRICULA 202504009
while True: #loop que dita a quantidade total de lançamentos financeiros que serão realizados ao longo do progama.
    try:
        lan = int(input("Digite quantos lançamentos financeiros serão feitos: ")) # converte o valor para um int e armazena na variavel lan
        print("")
        break #encerra o loop
    except ValueError: #se a conversão para int falhar e exibida uma mensagem de erro e o loop e reiniciado.
        print("Número invalido. Por favor tente novamente.")
        print("")
```

```python
# QUESTÃO 2C
# ALUNO: JOAO GABRIEL DA SILVA, MATRICULA 202504009

#criando lista
gast_despe_es = [] #lista de valor gasto na categoria Essensial.
gast_despe_la = [] #lista de valor gasto na categoria Lazer.
Essencial = [] #lista para armazenar as despesas essenciais
Lazer = []  #lista para armazenar as despesas de lazer

m = 0 #contador 

while m < lan: #loop while principal que se repete ate que todas os lançamentos financeiro (lan) seja lançados e classificados.
    try:
        #aqui o usuario deve escolher a categoria que sera feito o lançamento financeiro. Sendo que para seguir como uma classificação de Essencial e necessario que o usuario digite 1, para seguir como classificação Lazer e necessario que o usuario digite qualquer numero.
        caa = int(input("Em qual categoria sera lançada a despesa. Digite 1 para a categoria Essencial ou qualquer outro numero para a categoria Lazer: "))
        m += 1 #o contador aumenta sua contagem em 1
        print("")
        if caa == 1: #essa estrutura condicional verifica se o valor digitado anteriormente foi 1 
            es = str(input("Digite o nome da despesa: "))  #converte para string e armazena na variavel
            Essencial.append(es) #adiciona a despesa(string) na lista Essencial
            print("")
            while True: #loop que determina o valor gasto na despesa dita anteriormente
                try:
                    des = float(input(f"Digite o valor (em reais) que sera gasto: R$")) #converte o valor para float e armazena na variavel
                    des_a = round(des, 2) #arredonda o valor 
                    gast_despe_es.append(des_a) #adiciona o valor na lista de valor gasto em Essencial.
                    print("")
                    break #encerra este loop while e reinicia o loop principal. Levando o ususario ao inicio do processo de classificação. Mas se o contador tiver chegado no valor o suficiente para satisfazer a condição do loop while principal o loop principal e encerrado.
                except ValueError: # se a conversão para float falhar e exibida uma mensagem de erro e o loop e reiniciado.
                    print("Valor invalido. Por favor tente novamente.")
                    print("")
        else: #verifica se foi digitado qualquer outro numero que não seja igual a 1
            la = str(input("Digite o nome da despesa: ")) #converte para string e armazena na variavel
            Lazer.append(la) #adiciona a despesa(string) na lista Lazer
            print("")
            while True: #loop que determina o valor gasto na despesa dita anteriormente.
                try:
                    dla = float(input(f"Digite o valor (em reais) que sera gasto em: R$")) #converte o valor para float e armazena na variavel
                    dla_a = round(dla, 2) #arredonda o valor 
                    gast_despe_la.append(dla_a) #adiciona o valor na lista de valor gasto em Lazer.
                    print("")
                    break #encerra este loop while e reinicia o loop principal. Levando o ususario ao inicio do processo de classificação. Mas se o contador tiver chegado no valor o suficiente para satisfazer a condição do loop while principal o loop principal e encerrado.
                except ValueError: # se a conversão para float falhar e exibida uma mensagem de erro e o loop e reiniciado.
                    print("Valor invalido. Por favor tente novamente.")
                    print("")
    except ValueError: #se a conversão para int falhar e exibida uma mensagem de erro e o loop e reiniciado
        print("Valor invalido. Por favor tente novamente.")
        print("")
```

```python
# QUESTÃO 2D
# ALUNO: JOAO GABRIEL DA SILVA, MATRICULA 202504009

# extrutura condicional que:

if not gast_despe_es: #se a lista estiver vazia
    gast_despe_es.append(0) #adiciona o valor zero
if not gast_despe_la: #se a lista estiver vazia
    gast_despe_la.append(0) #adiciona o valor zero
if not Essencial: #se a lista estiver vazia
    Essencial.append(None) #adiciona um None na lista
if not Lazer: #se a lista estiver vazia
    Lazer.append(None) #adiciona um None na lista

valTot_e = sum(gast_despe_es) #soma todos os valores da lista de gastos na categoria Essencial
valTot_e_a = round(valTot_e, 2) #arredonda o valor

valTot_l = sum(gast_despe_la) #soma todos os valores da lista de gastos na categoria Lazer
valTot_l_a = round(valTot_l) #arredonda o valor

valTot = valTot_e_a + valTot_l_a #soma do valor total das listas dos gastos nas categorias Essencial e Lazer
valTot_l_aa = round(valTot, 2) #arredonda o valor

med_e = np.mean(gast_despe_es) #calcula a media dos valores da lista de gastos na categoria Essencial
med_e_a = round(med_e, 2) #arredonda o valor

med_l = np.mean(gast_despe_la) #calcula a media dos valores da lista de gastos na categoria Lazer
med_l_a = round(med_l, 2) #arredonda o valor


#exibindo os resultados de forma facil de entender.
print("="*40) #apenas para ajudar na separação das informações exibidas
print(f"Foram feitas {lan} despesas no mês com um valor total de R${valTot_l_aa}.")
print("")
print("="*40)
print(f"O gasto total na categoria essenciais foi de R${valTot_e_a}.")
print("")
print("="*40)
print(f"O gasto total na categoria lazer foi de R${valTot_l_a}")
print("")
print("="*40)
print(f"A média de valor em gastos essenciais foi de R${med_e_a}.")
print("")
print("="*40)
print(f"A média de valor em gastos com lazer foi de R${med_l_a}")
print("")
print("="*40)


maior_val_e = np.max(gast_despe_es) #localizar o valor mais alto na lista de gastos na categoria Essencial
maior_val_l = np.max(gast_despe_la) #localizar o valor mais alto na lista de gastos na categoria Lazer



if maior_val_e > maior_val_l: #verifica se o valor mais alto na lista de gastos na categoria Essencial e maior do que o valor mais alto na lista de gastos na categoria Lazer
    print(f"A categoria que possiu o maior valor gasto é Essencial com um gasto de R${maior_val_e}.") #exibe os resultado da verificação
    print("")
    print("="*40)

elif maior_val_l > maior_val_e: #verifica se o valor mais alto na lista de gastos na categoria Lazer e maior do que o valor mais alto na lista de gastos na categoria Essencial
    print(f"A categoria que possiu o maior valor gasto é Lazer com um gasto de R${maior_val_l}.") #exibe os resultado da verificação
    print("")
    print("="*40)

if valTot_e_a > valTot_l_a: #verifica se a soma de todos os valores na lista de gastos na categoria Essencial foi maior do que a soma de todos os valores na lista de gastos na categoria Lazer.
    print(f"A categoria que houve o maior gasto total foi a categoria Essencial com o valor gasto de R${valTot_e_a}.") #exibe os resultado da verificação
    print("")
    print("="*40)

elif valTot_l_a > valTot_e_a: #verifica se a soma de todos os valores na lista de gastos na categoria lazer foi maior do que a soma de todos os valores na lista de gastos na categoria Essencial.
    print(f"A categoria que houve o maior gasto total foi a categoria Lazer com o valor gasto de R${valTot_l_a}.") #exibe os resultado da verificação
    print("")
    print("="*40)

def class_or(sub_or,orcamento,categoria): # cria uma função que mostra se a meta foi alcançado ou não ou se foi utrapassada. os parametros dessa função são: subtração do orçamento pelo valor total gasto em uma categoria, o orçamento das categorias, nome da categoria.
    if sub_or < 0: #verifica se o valor obtido na subtração do orçamento pelo valor total gasto em uma categoria foi menor do que 0, ou seja, se foi um valor negativo.
        sub_or_p = abs(sub_or) #tranforma o valor negativo em um valor positivo
        return print(f"A meta de {categoria} que equivale a R${orcamento} foi ultrapassada. considere fazer um aumento de R${sub_or_p} ou diminuir as despesas.") #retorna esse reultado
    elif sub_or == 0: #verifica se o valor obtido na subtração do orçamento pelo valor total gasto em uma categoria foi igual a 0
        return print(f"A meta de {categoria} foi totalmente utilizada.") #retorna esse reultado
    elif sub_or > 0: #verifica se o valor obtido na subtração do orçamento pelo valor total gasto em uma categoria foi maior do que 0
        return print(f"Restam R${sub_or} do orçamento de R${orcamento} da categoria {categoria} para bater a meta.") #retorna esse reultado
    
sub_or_e = orçamento_es_a - valTot_e_a # o orçamento destinado a gastos na categoria Essencial e subtraido pelo valor total da lista do gasto na categoria Essencial
sub_or_e_a = round(sub_or_e, 2) #arredonda o valor

sub_or_l = orçamento_la_a - valTot_l_a # o orçamento destinado a gastos na categoria Lazer e subtraido pelo valor total da lista do gasto na categoria Lazer
sub_or_l_a = round(sub_or_l, 2) #arredonda o valor

class_or(sub_or_e_a,orçamento_es_a,"Essencial") #utilizamos a função criada anteriormente para mostrar se a meta da categoria Essencial foi alcançado ou não ou se foi utrapassada. E exibi o resultado
print("")
print("="*40)
class_or(sub_or_l_a,orçamento_la_a,"Lazer") #utilizamos a função criada anteriormente para mostrar se a meta da categoria lazer foi alcançado ou não ou se foi utrapassada. E exibe o resultado.
```
